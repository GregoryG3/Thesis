```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set(style="white", font_scale=0.8)

# Define the data for the confusion matrix
matrix_data = np.array([
    [7, 21, 4, 0],
    [7, 36, 7, 0],
    [0, 17, 67, 0],
    [7, 2, 1, 4]
])

# Calculate percentages
total = np.sum(matrix_data)
percentages = (matrix_data / total * 100).flatten()

# Combine counts and percentages in the labels
labels = np.array(["{0}\n{1:.1f}%".format(count, pct) if pct != 0 else "{0}\n0%".format(count)
                   for count, pct in zip(matrix_data.flatten(), percentages)]).reshape(4, 4)


# Create the heatmap
plt.figure(figsize=(5, 5)) #, dpi=200)
ax = sns.heatmap(matrix_data, annot=labels, fmt='', cmap='PuBu',  cbar=False,
                 xticklabels=['Unemotional', 'Somewhat\nEmotional', 'Rather\nEmotional', 'Very\nEmotional'],
                 yticklabels=['Unemotional', 'Somewhat\nEmotional', 'Rather\nEmotional', 'Very\nEmotional'])

ax.set_xticklabels(ax.get_xticklabels(), rotation=0, va='center', ha='center', fontsize=8)
ax.set_yticklabels(ax.get_yticklabels(), rotation=90, va='center', ha='center', fontsize=8)


# Title and labels
plt.title('DeBERTa V3: Text Emotionality Assessment', fontsize=10, fontweight='bold')
plt.xlabel('Predicted Label', fontweight='bold')
plt.ylabel('True Label', fontweight='bold')

# Add a border around the heatmap
for _, spine in ax.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(0.35)
    spine.set_color('black')

plt.savefig('cm_emotion.png', dpi=350)
plt.show()
```
